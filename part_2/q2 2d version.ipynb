# 2. Posegraph Optimisation


## 2.2: 2-D SLAM

In real-life application, we usually have wheel odometry and visual odometry data. Due to noise in wheel odometry it generates a noisy estimate of the trajectory. Our task is to use loop closure pairs to correct the drift. Visual odometry, due to heavy processing/computation need, is limited most of the time for getting loop closure constraints/edges. Remember we have 3 different kind of edges discussed so far:**Anchor, Odometry and loop closure edges**.

### TODO: 1
**TO-DO-1 :** Mention what each edge type is used for in 1-2 lines only.

**Odometry edges**

Odometry information describes the movement between poses at 2 consecutive timesteps.
Odometry inromation is usually represented as 2D or 3D transformations in SE(2) or in SE(3).
Such data might be collected using wheel encoders, GPS, IMU or other sophisticated approaches like visual odometry


**Loop closure edges**

Loop closure is the recognition of when the robot has returned to a previously mapped region and the use of this information to reduce uncertainty in the map estimate.
These constraints are generated by the place recognition algorithm
in the **front-end** and such constraints can also be between 2 poses which are noted in NON-consecutive timesteps.



**Anchor edges**

Since the entire pose-graph problem is modelled using relative transformations between the different poses, there can be many optinal solutions where all the relative transformations remain the same and the only change is the location of the first pose (at the first timestep). In order to remove this ambiguity, we generally tether the first pose to a particular location (eg: the origin) via the anchor based constraint.



# -------------------------------------------------------------------------------------


We pose this problem as a graph optimization problem. In our graph, poses are the vertices and constraints are the edges. To define vertex/orientation of bot we just need three values i.e. X, Y and THETA. Motion model can defined as:

$$x_{k+1} = x_{k} + \Delta x_{(k,k+1)} \cos(\theta_k) - \Delta y_{(k,k+1)} \sin(\theta_k) \\
y_{k+1} = y_{k} + \Delta y_{(k,k+1)} \cos(\theta_k) + \Delta x_{(k,k+1)} \sin(\theta_k) \\
\theta_{k+1} = \theta_{k}+  \Delta \theta_{(k,k+1)} \tag{3}$$

**Note:** Even the loop closure nodes are related by the above model. In case of loop-closure pairs it need not necessarily be consecutive nodes k and k+1.

We know that the initialization can be using odometry edges/loop closure oedges or BOTH.
In this case, I will initialize using `odometry edges`

````python
def prepare_initialization(vertices, edges):
    # sort vertices by index just to be sure
    vertices = copy.deepcopy(sorted(vertices, key=lambda x:x['vertex_id']))
    print("Number of vertices initially is ", len(vertices))
    NUM_VERTICES = max([max(x['v1'], x['v2']) for x in edges]) + 1
    print("Wanted number of vertices is ", NUM_VERTICES)
    
    for i in range(1, NUM_VERTICES):
        obj = dict()
        obj['vertex_id'] = i
        obj['x_c'] = 0
        obj['y_c'] = 0
        obj['theta_c'] = 0
        vertices.append(obj)
    
    # minor invariant check
    for idx, curr_ver in enumerate(vertices):
        assert(curr_ver['vertex_id']==idx)
        
    # topological sort
    edges = sorted(edges, key = lambda x:x['v1'])
    
    # go through all odometry edges and re-init all vertices
    for curr_edge in edges:
        v1 = curr_edge['v1']
        v2 = curr_edge['v2']
        if v2 == v1+1:
            
            #print(v2)
            # this is an odometry based edge
            x1 = vertices[v1]['x_c']
            y1 = vertices[v1]['y_c']
            theta1 = vertices[v1]['theta_c']
            
            dx = curr_edge['dx']
            dy = curr_edge['dy']
            dtheta = curr_edge['dtheta']
                       
            x2 = x1 + dx*np.cos(theta1) - dy * np.sin(theta1)
            y2 = y1 + dy*np.cos(theta1) + dx * np.sin(theta1)
            theta2 = theta1 + dtheta
            
            vertices[v2]['x_c'] , vertices[v2]['y_c'] , vertices[v2]['theta_c'] = x2, y2, theta2
            
            #print(x1, dx, x2, vertices[v2]['x_c'])
            #print("####")
    return vertices
````

````python
vertices = prepare_initialization(vertices, edges)
````

````python
def write_g2o_file(vertices, edges, write_path):
    
    with open(write_path, 'w') as fd:
        # write all vertices
        for curr_v in vertices:
            line_str = f"VERTEX_SE2 {curr_v['vertex_id']} {curr_v['x_c']} {curr_v['y_c']} {curr_v['theta_c']}\n"
            fd.write(line_str)
        
        
        # write FIX also
        fd.write("FIX 0\n")
        
        # write all edges
        for curr_edge in edges:
            line_str = f"EDGE_SE2"
            line_str += f" {curr_edge['v1']}"
            line_str += f" {curr_edge['v2']}"
            line_str += f" {curr_edge['dx']}"
            line_str += f" {curr_edge['dy']}"
            line_str += f" {curr_edge['dtheta']}"
            line_str += f" {curr_edge['x_conf']}"
            line_str += f" 0"
            line_str += f" 0"
            line_str += f" {curr_edge['y_conf']}"
            line_str += f" 0"
            line_str += f" {curr_edge['theta_conf']}"
            line_str+="\n"
            fd.write(line_str)       
````
